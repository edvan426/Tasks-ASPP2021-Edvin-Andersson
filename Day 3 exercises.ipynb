```python
# Ex 2 Day 3
# Program to multiply two matrices using nested loops
import numpy as np
import random

N = 250

# NxN matrix
X = np.random.randint(100, size=(N,N))

# Nx(N+1) matrix
Y = np.random.randint(100, size=(N,N+1))

# new method
result = np.matmul(X,Y)

# old method
resultold = []
for i in range(N):
    resultold.append([0] * (N+1))

# iterate through rows of X
for i in range(len(X)):
    # iterate through columns of Y
    for j in range(len(Y[0])):
        # iterate through rows of Y
        for k in range(len(Y)):
            resultold[i][j] += X[i][k] * Y[k][j]

# checking if same
print(np.where(result != resultold))
```

```python
# Ex 3 Day 3
class Person(object):
    def __init__(self, name, surname):
        self.name = name
        self.surname = surname
    def __str__(self):
        return "%s %s" % (self.name,self.surname)

class Student(Person):
    def __init__(self, name, surname, subject):
        Person.__init__(self, name, surname)
        self.subject = subject
    def __str__(self):
        return "%s %s, %s" % (self.name,self.surname,self.subject)

me = Person('Edvin','Andersson')
print(me)
measastudent = Student('Edvin','Andersson','Chemistry')
print(measastudent)
```
