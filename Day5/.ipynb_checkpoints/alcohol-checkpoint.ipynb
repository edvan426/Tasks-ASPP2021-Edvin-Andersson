### Step 1. Import the necessary libraries

```python
import pandas as pd
import io
import requests
import numpy as np
pd.set_option('display.max_rows', 500)
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv). 

### Step 3. Assign it to a variable called drinks.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv"
s = requests.get(url).content
drinks = pd.read_csv(io.StringIO(s.decode('utf-8')))
drinks['continent'].fillna('Unassigned', inplace=True)
print(drinks)
```

### Step 4. Which continent drinks more beer on average?

```python
conts = drinks['continent'].unique()
totbeer = pd.DataFrame()
for cont in conts:
    totbeer[cont] = [drinks['beer_servings'].loc[drinks['continent']==cont].sum()]

print('The',totbeer.idxmax(axis=1).to_string(index=False),'drinks the most beer')
```

### Step 5. For each continent print the statistics for wine consumption.

```python
for cont in conts:
    print('In',cont)
    print(drinks[['country','wine_servings']].loc[drinks['continent']==cont].to_string(index=False))
```

### Step 6. Print the mean alcoohol consumption per continent for every column

```python
meantotalcohol = pd.DataFrame()
meanwine = pd.DataFrame()
meanspirit = pd.DataFrame()
meanbeer = pd.DataFrame()
for cont in conts:
    meantotalcohol[cont] = [drinks['total_litres_of_pure_alcohol'].loc[drinks['continent']==cont].mean()]
    meanwine[cont] = [drinks['wine_servings'].loc[drinks['continent']==cont].mean()]
    meanspirit[cont] = [drinks['spirit_servings'].loc[drinks['continent']==cont].mean()]
    meanbeer[cont] = [drinks['beer_servings'].loc[drinks['continent']==cont].mean()]
print('Mean total litres of alcohol:')
print(meantotalcohol)
print('Mean total wine servings')
print(meanwine)
print('Mean total spirit servings')
print(meanspirit)
print('Mean total beer servings')
print(meanbeer)
```

### Step 7. Print the median alcoohol consumption per continent for every column

```python
mediantotalcohol = pd.DataFrame()
medianwine = pd.DataFrame()
medianspirit = pd.DataFrame()
medianbeer = pd.DataFrame()
for cont in conts:
    mediantotalcohol[cont] = [drinks['total_litres_of_pure_alcohol'].loc[drinks['continent']==cont].median()]
    medianwine[cont] = [drinks['wine_servings'].loc[drinks['continent']==cont].median()]
    medianspirit[cont] = [drinks['spirit_servings'].loc[drinks['continent']==cont].median()]
    medianbeer[cont] = [drinks['beer_servings'].loc[drinks['continent']==cont].median()]
print('median total litres of alcohol:')
print(mediantotalcohol)
print('median total wine servings')
print(medianwine)
print('median total spirit servings')
print(medianspirit)
print('median total beer servings')
print(medianbeer)
```

### Step 8. Print the mean, min and max values for spirit consumption.
