```python
#The four part subplot I did for the last day has one of the cleanest most structured plots I've ever made with matplotlib, so I'm gonna use it for this task as well
import numpy as np
import matplotlib.pyplot as plt
fig1, axs = plt.subplots(2,2, figsize=(15,10))

from scipy.stats import poisson

mu = 0.7
x1 = np.arange(0,5)
axs[0,0].plot(x1, poisson.pmf(x1, mu),'.', label='poisson pmf')
axs[0,0].plot(x1, poisson.cdf(x1, mu),'.', label='poisson cdf')
axs[0,0].legend()

drv = poisson.rvs(mu, size=1000)

axs[0,1].hist(drv)

from scipy.stats import norm

x2 = np.linspace(norm.ppf(0.01),norm.ppf(0.99), 100)
axs[1,0].plot(x2, norm.pdf(x2), label='norm pdf')
axs[1,0].plot(x2, norm.cdf(x2), label='norm pdf')
axs[1,0].legend()

crv = norm.rvs(size=1000)

axs[1,1].hist(crv)
print()
```
